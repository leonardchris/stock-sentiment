```python
#gets values for specific sid from dataset
xl= dataset[dataset.sid == 19654]
```

```python
#sorts dataset by date, newest first
xl_df = odo(xl.sort('asof_date',ascending=False), pd.DataFrame)
```

```python
#makes a new array for a specific range of the dataset
lista=[]
for i in range(360,420):
    lista.append(xl_df[xl_df['total_scanned_messages'] > 0].ix[i])
```
